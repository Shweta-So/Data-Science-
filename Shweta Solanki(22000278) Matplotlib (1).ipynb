```python
import pandas as pd
```

```python
data = [[50,56,89,40,50,85],
       [84,75,64,58,74,62],
       [55,65,74,80,52,40],
       [12,52,34,15,69,55],
       [25,40,33,10,25,60]]
columns=["Data Science","OOAD","python","Maths","Java","Statistics"]
index = ["Dev","Mahek","Diya","Suraj","Harsh"]
df = pd.DataFrame(data,columns = columns , index=index)
df
```

```python
fig, axes = plt.subplots(2,2,layout="constrained")

axes[0,0].plot(df.index,df.Maths,'m--',label='Math')
axes[0,0].set_xlabel('Names')
axes[0,0].set_ylabel('Maths Marks')
axes[0,0].set_title('Maths')
axes[0,0].legend()

axes[0,1].plot(df.index,df.python,'r--',label='py')
axes[0,1].set_xlabel('Names')
axes[0,1].set_ylabel('Python Marks')
axes[0,1].set_title('Python')
axes[0,1].legend()

axes[1,0].plot(df.index,df.OOAD,'b--',label='OOAD')
axes[1,0].set_xlabel('Names')
axes[1,0].set_ylabel('OOAD Marks')
axes[1,0].set_title('OOAD')
axes[1,0].legend()

axes[1,1].plot(df.index,df.Java,'y--',label='Java')
axes[1,1].set_xlabel('Names')
axes[1,1].set_ylabel('Java Marks')
axes[1,1].set_title('Java')
axes[1,1].legend()

fig.savefig("Figure.png")
```

```python
np.random.seed(3)
x=4+np.random.normal(0,2,24)
y=4+np.random.normal(0,2,len(x))
sizes=np.random.uniform(45,80,len(x))
colors=np.random.uniform(15,95,len(x))
fig , ax = plt.subplots()
ax.scatter(x , y , s=sizes , c=colors)
ax.set(xlim=(0,8) , ylim=(0,8))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.7,cmap='magma',marker='s',edgecolors='red')
plt.colorbar()
```
